```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import pprint
import json

# Import API key
from api_keys import g_key
```

```python
#configure gmaps

gmaps.configure(api_key = g_key)

#retrieve WeatherPy dataframe

weather_df = pd.read_csv("../WeatherPy/weather_data.csv")
weather_df
```

```python
#Create Heatmap for humidity using WeatherPy dataframe

places = weather_df[["Latitude","Longitude"]]
humidity = weather_df["Humidity"]
max_intensity = humidity.max()

fig = gmaps.figure(zoom_level = 2, center = [0,0])

heat_layer = gmaps.heatmap_layer(places, weights = humidity, max_intensity = 100, point_radius = 10)

fig.add_layer(heat_layer)
fig
```

Figure above shows that humidity is hiest near to the coast.

# Ideal Conditions

```python
#set ideal conditions
hot = 30
cold = 24
windspeed1 = 6
windspeed2 = 4
clouds = 0

ideal_df = weather_df[weather_df["Max Temperature (°C)"]<hot]
ideal_df = ideal_df[ideal_df["Max Temperature (°C)"]>cold]
ideal_df = ideal_df[ideal_df["Wind Speed"]<windspeed1]
ideal_df = ideal_df[ideal_df["Wind Speed"]>windspeed2]
ideal_df = ideal_df[ideal_df["Cloudiness"]==clouds]
ideal_df
```

```python
reduced_ideal = ideal_df[["City", "Longitude", "Latitude", "Country"]]

reduced_ideal.reset_index(drop = True, inplace=True)

reduced_ideal
```

```python
reduced_ideal["Hotel Name"] = ""
reduced_ideal["Hotel Rating"] = ""
reduced_ideal["Hotel Address"] = ""
reduced_ideal
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"

params = {"radius": 5000,
         "keyword": "hotel",
         "type": "lodging",
         "key": g_key}

for index, row in reduced_ideal.iterrows():
    lat = row["Latitude"]
    long = row["Longitude"]
    
    params["location"] = f"{lat},{long}"
    
    hotel_name = requests.get(base_url, params=params).json()
    
    
    try:
        reduced_ideal.loc[index, "Hotel Name"] = hotel_name["results"][0]["name"]
        reduced_ideal.loc[index, "Hotel Rating"] = hotel_name["results"][0]["rating"]
        reduced_ideal.loc[index, "Hotel Address"] = hotel_name["results"][0]["vicinity"]
    except (KeyError, IndexError):
        pass
    
```
